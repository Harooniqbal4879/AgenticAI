Next, you'll need to obtain an API key from OpenAI. If you don't have one, you can get it from the [OpenAI website](https://platform.openai.com/account/api-keys).

**Important:** For security, it's best practice *not* to hardcode your API key directly in your notebook. Instead, we'll store it securely using Colab's Secrets Manager. Click the "ðŸ”‘" icon on the left panel, then add your API key there, giving it the name `OPENAI_API_KEY`.

```python
from openai import OpenAI
from google.colab import userdata

# Get the API key from Colab's Secrets Manager
client = OpenAI(api_key=userdata.get('OPENAI_API_KEY'))
```

Now, let's create a simple prompt suitable for 9-12 graders and send it to the OpenAI API. We'll use the `gpt-3.5-turbo` model for this example.

Here's a prompt example: "Explain the concept of photosynthesis in simple terms, suitable for a 9th grader."

```python
prompt_text = "Explain the concept of photosynthesis in simple terms, suitable for a 9th grader."

response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": "You are a helpful assistant for high school students."},
        {"role": "user", "content": prompt_text}
    ]
)

# Display the answer
print("--- OpenAI's Answer ---")
print(response.choices[0].message.content)
print("----------------------")
```
